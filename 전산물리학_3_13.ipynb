```python
# 임시예제파일-감쇠진
import pylab as pl

time = pl.linspace(0, 4 * pl.pi, 1001)  # 0부터 4π까지 1001개의 균등한 시간 값 생성
height = pl.exp(- time / 3.0) * pl.sin(time * 3)  # 감쇠되는 사인파: 지수 감쇠 * 빠른 사인 진동

# pl.figure()

pl.plot(time, height, 'm-^')  # 감쇠 진동을 자홍색 -^로 그림
pl.plot(time, 0.3 * pl.sin(time * 3), 'g-')  # 일정 진폭의 사인파를 초록색-으로 그림

pl.legend(['damped', 'constant amplitude'])  # 범례 추가: 첫 번째는 감쇠, 두 번째는 일정 진폭
pl.xlabel('Time (s)')  # x축 라벨 설정
pl.ylabel('height')  # y축 라벨 설정

pl.show()
```

## 0.1 Libraries


*   파이썬 자체로 그래프를 그릴 수는 없다.
*   그래프를 그리기 위한 라이브러리가 필요하다.
- pylab: 수학적, 과학적 그래프 그리기
- matplotlib: 3차원 그래프 등 그리기



```python
#키르히호프 법칙을 이용한 식을 푸는 코딩
import numpy as np

# 계수 행렬 A 정의
A = np.matrix([[-13, 2, 4],
               [2, -11, 6],
               [4, 6, -15]])

# 상수항 벡터 B 정의 (방정식의 우변 값)
B = np.array([5, -10, 5])

# 선형 방정식 A * X = B를 풀어 해 X를 구함
C = np.linalg.solve(A, B)

print(C) #(예: 전류 I1, I2, I3)
```

## 01. Calculate

```python
# 다음을 확인하여라. (100% 출제)

#뺄셈과 곱셈: 정수로 출력
print(15-3)         # 12
print(4*7)          # 28

# 나눗셈: 소수점까지 실수로 출력
print(5/2)          # 2.5
print(4/2)          # 2.0

#실수와 정수의 뺄셈: 실수로 출력
print(17.0-10.0)    # 7.0
print(17.0-10)      # 7.0
print(17-10.0)      # 7.0
print(17-10.)       # 7.0
print(17.-10)       # 7.0


print(int(1.7))     # 1   소수를 int()를 씌울 때, 소수점 이하는 버림. 정수로 출력)
print(17//10)       # 1   (몫: 정수 나눗셈)
print(int(-1.7))    # -1  (소수점 이하 버림)

#나눈 값이 음수로 나올 때의 몫: 그냥 소수점 자리를 없애는 게 아니라, 더 작은 정수로 내려감
#몫은 기본적으로 내린다고 기억을 할것 (절댓값을 내리는게 아니라 작아진다)
print(-17//10)      # -2  (-1.7에서 내림)
print(17//-10)      # -2  (-1.7에서 내림)

#나머지: 나머지 %의 결과는 항상 나누는 수의 부호를 따라간다
print(53%24)        # 5   (53 = 24*2 + 5)
print(-17%10)       # 3   (-17 = 10*(-2) + 3)
print(17%-10)       # -3  (17 = (-10)*(-2) + (-3))
print(-17%-10)      # -7  (-17 = (-10)*1 + (-7))

print(3**4)         # 81  (3의 4제곱)
print(-5)           # -5
print(--5)          # 5   (음수 부호 두 번 → 양수)
print(---5)         # -5  (음수 부호 세 번 → 음수)

print(0.1+0.2)      # 0.30000000000000004 (부동소수점 오차)

if 0.1+0.2==0.3:
  print('세용바보') # (실행 안됨)
else:
  print('종원종원') # 종원종원
```
